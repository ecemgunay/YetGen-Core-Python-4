```python
#İleri Birleştirme (Merge işlemleri)
df1 = pd.DataFrame({"Çalışanlar": ["Ecem", "Sezer", "Ceren", "Özlem"],
                    "Grup" : ["Koordinatör", "İş Geliştirme", "Full Time Çalışan", "Eğitim Koordinatörü"]})
df1
```

```python
df2 = pd.DataFrame({"Çalışanlar": ["Ecem", "Sezer", "Ceren", "Özlem"],
                     "İlk Giriş Yılı":[2017,2019,2022,2018]})
df2
```

```python
pd.merge(df1,df2)
```
