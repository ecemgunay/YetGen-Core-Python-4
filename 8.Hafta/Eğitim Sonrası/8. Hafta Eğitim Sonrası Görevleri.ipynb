# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
#VKİ nasıl hesaplanır: vücut ağırlığının, boyun karesine bölünmesiyle hesaplanır.
class Kullanici:
    def __init__(self, isim_soyisim, yas, boy, kilo):
        self.isim_soyisim = isim_soyisim
        self.yas = yas
        self.boy = boy
        self.kilo = kilo
        self.vki = self.vki_hesapla()
        self.sinif = self.vki_siniflandir()
    
    def vki_hesapla(self):
        return self.kilo / (self.boy ** 2)
    
    def vki_siniflandir(self):
        if self.vki < 18.5:
            return "Zayıf"
        elif 18.5 <= self.vki < 24.9:
            return "Normal"
        elif 25 <= self.vki < 29.9:
            return "Fazla Kilolu"
        elif 30 <= self.vki < 39.9:
            return "Obez"
        else:
            return "İleri Derecede Obez (morbid obez)"
    
    def bilgi_satiri(self):
        return f"{self.isim_soyisim}, Yaş: {self.yas}, Boy: {self.boy}, Kilo: {self.kilo}, VKİ: {round(self.vki, 2)}, Sınıf: {self.sinif}"


def main():
    print("Merhaba, VKİ Hesaplama Programına Hoş Geldiniz!")
    print("Kaç kişi için giriş yapmak istersiniz?")
    kac_kisi = int(input("Kişi sayısını giriniz: "))

    kullanicilar = []

    for i in range(kac_kisi):
        print(f"\n{i+1}. kişi bilgileri:")
        ad_soyad = input("İsim Soyisim: ")
        yas = int(input("Yaş: "))
        boy = float(input("Boy (metre): "))
        kilo = float(input("Kilo (kg): "))
        
        kullanici = Kullanici(ad_soyad, yas, boy, kilo)
        kullanicilar.append(kullanici)

    with open("vki_verileri.txt", "w", encoding="utf-8") as file:
        for k in kullanicilar:
            file.write(k.bilgi_satiri() + "\n")

    print("\n📁 Tüm veriler 'vki_verileri.txt' dosyasına kaydedildi.")


if __name__ == "__main__":
    main()


        
```
