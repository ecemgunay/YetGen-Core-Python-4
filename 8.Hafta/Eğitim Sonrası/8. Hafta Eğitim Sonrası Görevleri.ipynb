SÄ±nÄ±flar ise ÅŸu ÅŸekilde: 
* 18, 5 kg/m.'nin altÄ±nda olanlar: ZayÄ±f
* 18.5 â€“ 24, 9 kg/m. arasÄ±nda olanlar: Normal kilolu
* 25 â€“ 29, 9 kg/m. arasÄ±nda olanlar: Fazla kilolu
* 30 â€“ 39, 9 kg/m. arasÄ±nda olanlar: Obez
* 40 kg/m.'nin Ã¼zerinde olanlar: Ä°leri derecede obez (morbid obez), olarak gÃ¶rÃ¼lÃ¼r.

```python
# 1.Problemin Ã‡Ã¶zÃ¼mÃ¼nÃ¼ Buraya YazÄ±nÄ±z
#VKÄ° nasÄ±l hesaplanÄ±r: vÃ¼cut aÄŸÄ±rlÄ±ÄŸÄ±nÄ±n, boyun karesine bÃ¶lÃ¼nmesiyle hesaplanÄ±r.
class Kullanici:
    def __init__(self, isim_soyisim, yas, boy, kilo):
        self.isim_soyisim = isim_soyisim
        self.yas = yas
        self.boy = boy
        self.kilo = kilo
        self.vki = self.vki_hesapla()
        self.sinif = self.vki_siniflandir()
    
    def vki_hesapla(self):
        return self.kilo / (self.boy ** 2)
    
    def vki_siniflandir(self):
        if self.vki < 18.5:
            return "ZayÄ±f"
        elif 18.5 <= self.vki < 24.9:
            return "Normal"
        elif 25 <= self.vki < 29.9:
            return "Fazla Kilolu"
        elif 30 <= self.vki < 39.9:
            return "Obez"
        else:
            return "Ä°leri Derecede Obez (morbid obez)"
    
    def bilgi_satiri(self):
        return f"{self.isim_soyisim}, YaÅŸ: {self.yas}, Boy: {self.boy}, Kilo: {self.kilo}, VKÄ°: {round(self.vki, 2)}, SÄ±nÄ±f: {self.sinif}"


def main():
    print("Merhaba, VKÄ° Hesaplama ProgramÄ±na HoÅŸ Geldiniz!")
    print("KaÃ§ kiÅŸi iÃ§in giriÅŸ yapmak istersiniz?")
    kac_kisi = int(input("KiÅŸi sayÄ±sÄ±nÄ± giriniz: "))

    kullanicilar = []

    for i in range(kac_kisi):
        print(f"\n{i+1}. kiÅŸi bilgileri:")
        ad_soyad = input("Ä°sim Soyisim: ")
        yas = int(input("YaÅŸ: "))
        boy = float(input("Boy (metre): "))
        kilo = float(input("Kilo (kg): "))
        
        kullanici = Kullanici(ad_soyad, yas, boy, kilo)
        kullanicilar.append(kullanici)

    with open("vki_verileri.txt", "w", encoding="utf-8") as file:
        for k in kullanicilar:
            file.write(k.bilgi_satiri() + "\n")

    print("\nğŸ“ TÃ¼m veriler 'vki_verileri.txt' dosyasÄ±na kaydedildi.")


if __name__ == "__main__":
    main()


        
```
