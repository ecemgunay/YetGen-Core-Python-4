Kumanda isminde bir sÄ±nÄ±f oluÅŸturarak aÅŸaÄŸÄ±daki iÅŸlemleri yapmasÄ±nÄ± saÄŸlayan fonksiyonlarÄ± sÄ±nÄ±f iÃ§erisinde yazalÄ±m.
1. Tv AÃ§
2. Tv Kapat
3. Ses AyarlarÄ±
4. Kanal Ekle
5. Kanal SayÄ±sÄ±nÄ± Ã–ÄŸrenme
6. Rastgele Kanala GeÃ§me
7. Televizyon Bilgileri


KumandayÄ± kapatmak iÃ§in q tuÅŸuna basÄ±lmalÄ± ve diÄŸer durumlarda hangi iÅŸlemi yapmak istediÄŸi sorulmalÄ±dÄ±r. Her iÅŸlemde gerekli fonksiyonunun sÄ±nÄ±f iÃ§erisinden Ã§aÄŸÄ±rÄ±lmasÄ± gerekmektedir.

random ve time kÃ¼tÃ¼phanesini araÅŸtÄ±rarak kullanabilirsiniz.

Kendi istediÄŸiniz Ã¶zellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Ã‡Ã¶zÃ¼mÃ¼nÃ¼ Buraya YazÄ±nÄ±z
import random #Rastgele seÃ§imler yapmak iÃ§in gerekli
import time #Bekleme ve zamanlama iÅŸlemleri iÃ§in gerekli

class Kumanda:
    def __init__(self):
        self.tv_acik_mi = False #Televizyon aÃ§Ä±k mÄ± kontrolÃ¼
        self.ses = 10 #BaÅŸlangÄ±Ã§ sesi
        self.kanallar = ["Halk TV", "SÃ¶zcÃ¼ TV", "KRT", "TELE 1", "DW TÃ¼rkÃ§e"]
        self.mevcut_kanal = self.kanallar[0] #BaÅŸlangÄ±Ã§ kanalÄ±
        self.favori_kanallar = [] #Favori kanallar listesi
        self.son_kanal = None #Son izlenen kanal
    def tv_ac(self):
        if self.tv_acik_mi:
            print("Televizyon zaten aÃ§Ä±k.")
        else:
            print("Televizyon aÃ§Ä±lÄ±yor...")
            time.sleep(1)
            self.tv_acik_mi = True
            print("Televizyon aÃ§Ä±ldÄ±.")

    def tv_kapat(self):
        if not self.tv_acik_mi:
            print("Televizyon zaten kapalÄ±.")
        else:
            print("Televizyon kapanÄ±yor...")
            time.sleep(1)
            self.tv_acik_mi = False
            print("Televizyon kapandÄ±.")
    def ses_ayarlari(self):
        while True:
            islem = input("Sesi artÄ±r: > | azalt: < | Ã§Ä±k: q : ")
            if islem == "<":
                if self.ses < 100:
                    self.ses += 1
                    print(f"Ses: {self.ses}")
                else:
                    print("Ses zaten maksimumda. ")
            elif islem == ">":
                if self.ses >0:
                    self.ses -= 1
                    print(f"Ses: {self.ses}")
                else:
                    print("Ses zaten minimumda.")
            elif islem == "q": 
                print("Ses ayarÄ±ndan Ã§Ä±kÄ±lÄ±yor.")
                break
            else:
                print("GeÃ§ersiz iÅŸlem. LÃ¼tfen tekrar deneyin.")

    def kanal_ekle(self):
        yeni_kanal = input("Eklemek istediÄŸiniz kanalÄ±n adÄ±nÄ± girin: ")
        if yeni_kanal not in self.kanallar: # not in "iÃ§inde deÄŸilse" anlamÄ±na gelir. Yani self.kanallar listesinde yeni_kanal var mÄ± yok mu kontrolÃ¼ yapÄ±lÄ±r.
            self.kanallar.append(yeni_kanal)
            print(f"{yeni_kanal} kanalÄ± eklendi.")       
        else:
            print("Bu kanal zaten listede")
    def kanal_sil(self):
        silinecek_kanal = input("Silmek istediÄŸiniz kanalÄ±n adÄ±nÄ± girin: ")
        if silinecek_kanal in self.kanallar:
            self.kanallar.remove(silinecek_kanal)
            print(f"{silinecek_kanal} kanalÄ± silindi.")
        else:
            print("Bu kanal listede yok.")
    
    def kanal_sayisi(self):
        print(f"Toplam kanal sayÄ±sÄ±: {len(self.kanallar)}")

    def rastgele_kanal(self):
        self.son_kanal = self.mevcut_kanal # Mevcut kanalÄ± son kanala atÄ±yoruz.
        self.mevcut_kanal = random.choice(self.kanallar) #kanal listesinden rastgele bir kanal seÃ§ip, onu mevcut kanal (yani ÅŸuanda aÃ§Ä±k olan kanal) olarak atar.
        print(f"Son izlenen Kanal: {self.mevcut_kanal}")

    def televizyon_bilgileri(self):
        print("ğŸ“º Televizyon Bilgileri")
        print(f"AÃ§Ä±k mÄ±?: {'Evet' if self.tv_acik_mi else 'HayÄ±r'}")
        print(f"Ses: {self.ses}") #bu satÄ±r direkt olarak TV'nin mevcut ses dÃ¼zeyini yazdÄ±rÄ±r. Ã–rn: Ses: 14
        print(f"Mevcut Kanal: {self.mevcut_kanal}") #Bu da ÅŸu anda izlenen kanalÄ± gÃ¶sterir. 
        print(f"Kanallar: {', '.join(self.kanallar)}") # .join() fonksiyonu ile kanal listesi araya virgÃ¼l koyarak birleÅŸtirilir. 
        #Ã–rn: ["KRT", "TELE 1", "Halk TV"] -> "KRT, TELE 1, Halk TV" ÅŸeklinde yazdÄ±rÄ±r.
        print(f"Favoriler: {', '.join(self.favori_kanallar) if self.favori_kanallar else 'HenÃ¼z yok'}") #Favori kanallar listesi boÅŸsa "HenÃ¼z yok" yazdÄ±rÄ±r.
        #EÄŸer boÅŸ deÄŸilse, favori kanallar listesi araya virgÃ¼l koyarak birleÅŸtirilir.
        print(f"Son Ä°zlenen Kanal: {self.son_kanal if self.son_kanal else 'HenÃ¼z yok'}")
        #EÄŸer daha Ã¶nce kanal deÄŸiÅŸtirildiyse self.son_kanal deÄŸiÅŸkeni dolar ve onu gÃ¶sterir.
        #EÄŸer hiÃ§ kanal deÄŸiÅŸikliÄŸi yapÄ±lmadÄ±ysa "HenÃ¼z yok" yazdÄ±rÄ±r.

    def favoriye_ekle(self):
        kanal = input("Favoriye eklemek istediÄŸiniz kanal: ")
        if kanal in self.kanallar and kanal not in self.favori_kanallar: 
            #kanal in self.kanallar: girilen kanal gerÃ§ekten listede var mÄ±?
            #kanal not in self.favori_kanallar: girilen kanal zaten favorilerde var mÄ±?
            #EÄŸer her iki koÅŸul da saÄŸlanÄ±yorsa, kanalÄ± favorilere ekle.
            self.favori_kanallar.append(kanal)
            print(f"{kanal} favorilere eklendi.")
        else:
            print("Kanal mevcut deÄŸil ya da zaten favorilerde.")

    def kanal_listesi_alfabetik(self):
        print("Kanallar (A-Z)")
        for kanal in sorted(self.kanallar):
            print(f"- {kanal}")

#Ana kontrol paneli
kumanda = Kumanda()

while True:
    print("""
    1. Tv AÃ§
    2. Tv Kapat
    3. Ses AyarlarÄ±
    4. Kanal Ekle
    5. Kanal Sil
    6. Kanal SayÄ±sÄ±
    7. Rastgele Kanala GeÃ§
    8. Televizyon Bilgileri
    9. Favori Kanal Ekle
    10. Kanal Listesini Alfabetik GÃ¶ster
    Ã‡Ä±kmak iÃ§in q tuÅŸuna basÄ±nÄ±z.
    """)

    secim = input("Ä°ÅŸlem SeÃ§in: ")

    if secim == "1":
        kumanda.tv_ac()
    elif secim == "2":
        kumanda.tv_kapat()
    elif secim == "3":
        kumanda.ses_ayarlari()
    elif secim == "4":
        kumanda.kanal_ekle()
    elif secim == "5":
        kumanda.kanal_sil()
    elif secim == "6":
        kumanda.kanal_sayisi()
    elif secim == "7":
        kumanda.rastgele_kanal()
    elif secim == "8":
        kumanda.televizyon_bilgileri()
    elif secim == "9":
        kumanda.favoriye_ekle()
    elif secim == "10":
        kumanda.kanal_listesi_alfabetik()
    elif secim.lower() == "q":
        print("Kumanda kapatÄ±lÄ±yor...")
        break
    else:
        print("GeÃ§ersiz seÃ§im. LÃ¼tfen tekrar deneyin.")
```
